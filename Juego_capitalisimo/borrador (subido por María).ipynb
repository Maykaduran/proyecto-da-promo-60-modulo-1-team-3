```python
import random

capitalisimo = {
    "España": "Madrid",
    "Francia": "París",
    "Alemania": "Berlín",
    "Italia": "Roma",
    "Japón": "Tokio",
    "Brasil": "Brasilia" 
}

capitales_mundo = list(capitalisimo.values()) 

lista_todas_capitales_normalizadas = []
for capital in capitales_mundo:
    capital_limpia = capital.strip().lower()
    lista_todas_capitales_normalizadas.append(capital_limpia)


def iniciar_juego_simple():
    
    PUNTOS = 10 
    RONDAS = 5 
    
    print("=" * 35)
    print(f" JUEGO DE CAPITALES. Rondas: {RONDAS}")
    print(f" Puntos iniciales: {PUNTOS}")
    print("=" * 35)

    for ronda in range(1, RONDAS + 1):
        
        if PUNTOS <= 0:
            print("\n❌ ¡PUNTUACIÓN CERO! Juego terminado.")
            break
            
        print(f"\n--- RONDA {ronda} --- Puntos: {PUNTOS} ---")
        
        pais_elegido = random.choice(list(capitalisimo.keys()))
        capital_correcta = capitalisimo[pais_elegido]
        correcta_normalizada = capital_correcta.strip().lower()

        pregunta = f"¿Capital de {pais_elegido}?"
        respuesta_usuario = input(pregunta + " ").strip().lower()
        
        if respuesta_usuario == correcta_normalizada:
            print('✅ ¡Correcto!')
            
        else:
            if respuesta_usuario in lista_todas_capitales_normalizadas:
                PUNTOS -= 2
                print(f'❌ Incorrecto. -2 puntos.')
                
            else:
                PUNTOS -= 3
                print(f'❓ Respuesta absurda. -3 puntos.')
            
            print(f'   La capital correcta era: {capital_correcta}')

    print("\n" + "=" * 35)
    print(" JUEGO FINALIZADO")
    print(f" Puntuación Total: {PUNTOS}")
    print("=" * 35)

iniciar_juego_simple()
```
